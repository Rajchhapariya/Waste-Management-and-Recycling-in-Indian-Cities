```python
# Splitting the column "Landfill Location (Lat, Long)" because it has two 
df[["Latitude", "Longitude"]] = df["Landfill Location (Lat, Long)"].str.split(",", expand=True)

# Convert to numeric (remove spaces if any)
df["Latitude"] = pd.to_numeric(df["Latitude"].str.strip())
df["Longitude"] = pd.to_numeric(df["Longitude"].str.strip())

print(df.head())
```
